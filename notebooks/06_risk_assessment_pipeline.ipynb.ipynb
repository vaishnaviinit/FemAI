```python
import pandas as pd
import numpy as np
import joblib
```

```python
model = joblib.load(
    "../models/pcos_xgboost.pkl"
)

print("Model Loaded Successfully!")
```

```python
df = pd.read_csv(
    "../data/processed/cleaned_pcos.csv"
)

X = df.drop(
    "PCOS (Y/N)",
    axis=1
)

feature_names = X.columns.tolist()

print(len(feature_names))
```

```python
patient = X.iloc[[0]]

patient
```

```python
prediction = model.predict(patient)

probability = model.predict_proba(patient)

pcos_probability = probability[0][1] * 100

print(prediction)
print(pcos_probability)
```

```python
if pcos_probability < 30:
    risk_level = "Low"

elif pcos_probability < 70:
    risk_level = "Moderate"

else:
    risk_level = "High"

print(risk_level)
```

```python
recommendations = []

if patient["Weight gain(Y/N)"].values[0] == 1:
    recommendations.append(
        "Monitor weight and maintain a balanced diet."
    )

if patient["Fast food (Y/N)"].values[0] == 1:
    recommendations.append(
        "Reduce fast food consumption."
    )

if patient["Reg.Exercise(Y/N)"].values[0] == 0:
    recommendations.append(
        "Increase physical activity."
    )

recommendations
```

```python
print("=" * 50)

print("FEMAI PCOS RISK ASSESSMENT REPORT")

print("=" * 50)

print(
    f"Predicted PCOS Risk: {pcos_probability:.2f}%"
)

print(
    f"Risk Category: {risk_level}"
)

print("\nRecommendations:")

for rec in recommendations:
    print("-", rec)

print("=" * 50)
```

```python
report = {
    "Risk Score": round(
        pcos_probability,
        2
    ),
    "Risk Category": risk_level,
    "Recommendations": recommendations
}

report_df = pd.DataFrame([report])

report_df
```
