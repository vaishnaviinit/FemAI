```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/raw/PCOS_data.csv")

print(df.shape)
```

```python
columns_to_drop = [
    "Sl. No",
    "Patient File No.",
    "Unnamed: 44"
]

df = df.drop(columns=columns_to_drop)

print("New Shape:", df.shape)
```

```python
df.isnull().sum().sort_values(ascending=False).head(10)
```

```python
df["Marraige Status (Yrs)"] = (
    df["Marraige Status (Yrs)"]
    .fillna(df["Marraige Status (Yrs)"].median())
)

df["Fast food (Y/N)"] = (
    df["Fast food (Y/N)"]
    .fillna(df["Fast food (Y/N)"].mode()[0])
)
```

```python
print("Total Missing Values:", df.isnull().sum().sum())
```

```python
df.select_dtypes(include="object").columns
```

```python
df["II    beta-HCG(mIU/mL)"] = pd.to_numeric(
    df["II    beta-HCG(mIU/mL)"],
    errors="coerce"
)

df["AMH(ng/mL)"] = pd.to_numeric(
    df["AMH(ng/mL)"],
    errors="coerce"
)
```

```python
df.select_dtypes(include="object").columns
```

```python
corr_with_target = (
    df.corr(numeric_only=True)["PCOS (Y/N)"]
    .sort_values(ascending=False)
)

corr_with_target.head(15)
```

```python
df.to_csv(
    "../data/processed/cleaned_pcos.csv",
    index=False
)

print("Cleaned dataset saved successfully!")
```
