```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    classification_report
)

from xgboost import XGBClassifier

import joblib
```

```python
df = pd.read_csv("../data/processed/cleaned_pcos.csv")

df.columns = df.columns.str.strip()

print(df.shape)
```

```python
important_features = [
    "Age (yrs)",
    "Weight (Kg)",
    "BMI",
    "Cycle(R/I)",
    "Weight gain(Y/N)",
    "hair growth(Y/N)",
    "Skin darkening (Y/N)",
    "Fast food (Y/N)",
    "Reg.Exercise(Y/N)",
    "AMH(ng/mL)",
    "Follicle No. (L)",
    "Follicle No. (R)"
]

X = df[important_features]

y = df["PCOS (Y/N)"]

print(X.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print(X_train.shape)
print(X_test.shape)
```

```python
xgb_lite = XGBClassifier(
    n_estimators=200,
    max_depth=4,
    learning_rate=0.1,
    random_state=42
)

xgb_lite.fit(X_train, y_train)

print("Lite Model Trained Successfully!")
```

```python
y_pred = xgb_lite.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))

print("Precision:", precision_score(y_test, y_pred))

print("Recall:", recall_score(y_test, y_pred))

print("F1:", f1_score(y_test, y_pred))
```

```python
print(classification_report(
    y_test,
    y_pred
))
```

```python
comparison = pd.DataFrame({
    "Model": [
        "XGBoost Full (41 Features)",
        "XGBoost Lite (12 Features)"
    ],
    "Accuracy": [
        0.9266,
        accuracy_score(y_test, y_pred)
    ],
    "F1 Score": [
        0.8788,
        f1_score(y_test, y_pred)
    ]
})

comparison
```
